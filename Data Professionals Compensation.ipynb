```python
#Importing Libraries
import pandas as pd
import numpy as np
import os
from collections import Counter
#pd.set_option('display.max_rows', 2500)
#pd.set_option('display.max_columns', 50)
```

```python
#Importing Data
df=pd.read_csv(r"C:\Users\USER\Desktop\Personal\Data Professionals Compensation\Data Professional Compensation Data.csv")
df
```

```python
#Renaming columns
df.rename(columns= {"What is your Current Role?": "Occupation",
                    "If other, please indicate your role here":"Other Role",
                    "What is your Level?":"Level",
                    "How many years of experience do you have in the data Field?":"Years of Experience",
                    'What industry is  your Employer? eg Fintech, Utilities, HR, Gaming, Health etc.':"Industry",
                    "What is your gender":"Gender",
                    "What is your main Tech stack?":"Tech Stack",
                    "If other, please indicate your tool stack here":"Other Tech Stack",
                    "What is your monthly Gross Salary in Kes per month?":"Gross Salary(ksh)",
                    "Any other benefits?":"Other Benefits",
                    "What is your work setup?":"Work Setup"}, inplace= True)
```

```python
#Cleaning Role Column
df["Occupation"]= df["Occupation"].replace({"Other":" "})
df['Other Role'] = df['Other Role'].fillna('')


#Merging the columns to have "Role" column
df["Role"]= df["Occupation"]+df["Other Role"]

# Removing white space
df["Role"] = df["Role"].str.strip()

#Cleaning names- Removing typos and categorizing data
df["Role"]= df["Role"].replace({'Data Manager (Digital)':'Data Manager',
                    'Data AnalystManufacturing':'Data Analyst',
                    'Data AnalystAdvisory':'Data Analyst',
                    'data officer':'Data Officer',
                    'community manager':'Community Manager',
                    'HRM': 'Human Resources Manager',
                    'fulfilment supervisor':'Fulfilment Supervisor',
                    'Data AnalystRating Analyst':'Data Analyst',
                    'Business intelligence analyst':'Business Intelligence analyst',
                    'Risk Data Analyst':'Data Analyst',
                    'ERP Consultnat':'ERP Consultant',
                    'Data EngineerData Analytics Engineer':'Data Analytics Engineer',
                    'Inventory analyst':'Inventory Analyst',
                    'Data ScientistN/a':'Data Scientist',
                    'Data AnalystMEARLO':'Data Analyst',
                    'Data ScientistStudent':'Data Scientist Student',
                    'Software engineer': 'Software Engineer',
                    'Data AnalystJunior Data analyst': 'Data Analyst',
                    'Data AnalystFlight Data Analyst': 'Data Analyst',
                    'Data AnalystResearch Assistant in Water Resources':'Data Analyst',
                    'Business Central developer':'Business Central Developer',
                    'Data officer':'Data Officer',
                    'Data Mangement Associate':'Data Management Associate',
                    'supply planner':'Supply Planner',
                    '':'Data Analyst'
                    })           
```

```python
#Cleaning "level" Column
df['Level']=df['Level'].replace({'Junior  eg Junior Data Analyst': 'Junior',
                     'Mid-Level  eg Data Analyst': 'Mid-Level',
                    'Lead eg Lead Analyst': 'Lead',
                    'Senior Level  eg Senior Data Analyst': 'Senior Level',
                    'Manager eg Manager of Analytics': 'Manager'})
```

```python
#Cleaning "Industry" column
#Removing white space
df["Industry"]= df["Industry"].str.strip()
#Filling null with an empty string
df["Industry"] = df["Industry"].fillna('')
#Cleaning typos.
df["Industry"]= df["Industry"].replace({ 'Telco': 'Telecommunications',
                        'Fintech': 'FinTech',
                        'health': 'Health',
                        'Fast Moving Consumer Goods': 'FMCG',
                        'Fintech and IT': 'FinTech',
                        'supply chain': 'Supply Chain',
                        'Non Profit Organization': 'Non-Profit',
                        'Non-profit':'Non-Profit',
                        'Technology and Telecommunications':'Telecommunications',
                        'Market research': 'Market Research',
                        'Fmcg' : 'FMCG',
                        'Healtg': 'Health',
                        'Analytics->I work for a company in Europe not Kenya': 'Analytics',
                        'FMCG consulting': 'FMCG',
                        'Still searching': 'Unemployed',
                        'Non Profit': 'Non-Profit',
                        'Finetech': 'FinTech',
                        'Manufacturing (FMCG)':  'FMCG', 
                        'Hr':'HR',
                        'I do data analysis for different companies.': 'Consulting',
                        'Manufacturing, FMCGs': 'FMCG',
                        'machine learning': 'Machine Learning',       
                        'Nill': 'N/A',
                        'Nill': 'Nil',
                        'Finance - Pensions': 'Finance',
                        'Financial':  'Finance'})
```

```python
#Cleaning 'Tech Stack' column.
df['Tech Stack']= df['Tech Stack'].str.replace('Other','')
df['Other Tech Stack']= df['Other Tech Stack'].fillna('')
df['Tech Stacks']= df['Tech Stack']+" "+ df['Other Tech Stack']
```

```python
#Gross Salary column data cleaning
df['Gross Salary(ksh)']= df['Gross Salary(ksh)'].fillna(0)
df['Gross Salary(ksh)']= df['Gross Salary(ksh)'].replace(np.nan, 0)
df['Gross Salary(ksh)']=df['Gross Salary(ksh)'].str.replace(',','')
df['Gross Salary(ksh)']= df['Gross Salary(ksh)'].replace({'500$':65000, 
                                 '15000/-':15000,
                                 '112000KES':112000,
                                 '920 EUROS':129000,
                                 'Unemployed': 0,
                                 'Nill':0,
                                 '25': 25000,
                                 '600 USD (varies with exchange rates)': 78000,
                                 '2.5 million':2500000,
                                 '80000kshs': 80000,
                                 'ksh 60000': 60000,
                                 '':0,
                                 '00':0,
                                 '0':0,
                                 '35728.3': 35728})

#Dropping rows where Gross Salary is Zero and Null
df = df[(df['Gross Salary(ksh)'] != 0) & (~df['Gross Salary(ksh)'].isna())]
# Reseting Index
df.reset_index(drop=True, inplace=True)

#Dropping specific rows in the dataset with outliers
delete_rows = [109, 139, 141, 211]
df = df.drop(delete_rows)

#Changing data type to integer
df['Gross Salary(ksh)'] = df['Gross Salary(ksh)'].astype(int)
```

```python
#Cleaning 'Work Setup' column
df['Work Setup']=df['Work Setup'].replace({'Full remote':'Full Remote',
                              'Hybrid- I need to be in the office specific days in a week': 'Hybrid- Office specific days in a week',
                              'Hybrid-I choose my days to go the office': 'Hybrid-Choose days to go the office'})   
#Dropping columns
df= df.drop( columns= ['Timestamp', 'Occupation', 'Other Role', 'Tech Stack', 'Other Tech Stack'])
df
```

```python
#EDA
#Filtering data to have Data Analyst, Data Scientist, Data Engineer
df_filtered = df[df['Role'].isin(['Data Analyst', 'Data Scientist', 'Data Engineer'])]
#No of Respondent
No_of_Respondent= df_filtered.groupby('Role')["Role"].count().sort_values(ascending=False).reset_index(name="No of Respondent")
No_of_Respondent
```

```python
#Average Salaries
average_salaries= df_filtered.groupby('Role')['Gross Salary(ksh)'].mean().astype(int).sort_values(ascending=False).reset_index(name= 'Average Salary')
average_salaries
```

```python
#Data Scientist Average Salaries by Level
df_data_analyst_filtered = df[df['Role']=='Data Analyst']
data_analyst_average_salaries = df_data_analyst_filtered.groupby(['Role', 'Level'])['Gross Salary(ksh)'].mean().astype(int).reset_index(name='Average Salary').sort_values(by=['Role','Average Salary'], ascending=[False,True])
data_analyst_average_salaries
```

```python
#Data Scientist Average Salaries by Level
df_data_analyst_filtered = df[df['Role']=='Data Scientist']
data_scientist_average_salaries = df_data_analyst_filtered.groupby(['Role', 'Level'])['Gross Salary(ksh)'].mean().astype(int).reset_index(name='Average Salary').sort_values(by=['Role','Average Salary'], ascending=[False,True])
data_scientist_average_salaries
```

```python
#Data Engineers Average Salaries by Level
df_data_analyst_filtered = df[df['Role']=='Data Engineer']
data_engineers_average_salaries = df_data_analyst_filtered.groupby(['Role', 'Level'])['Gross Salary(ksh)'].mean().astype(int).reset_index(name='Average Salary').sort_values(by=['Role','Average Salary'], ascending=[False,True])
data_engineers_average_salaries
```

```python
#Data Analyst Average Salaries by Years of Experience
df_data_analyst_filtered = df[df['Role']=='Data Analyst']
data_analysts_average_salaries_by_years = df_data_analyst_filtered.groupby(['Role', 'Years of Experience'])['Gross Salary(ksh)'].mean().astype(int).reset_index(name='Average Salary').sort_values(by=['Role','Average Salary'], ascending=[False,True])
data_analysts_average_salaries_by_years 
```

```python
#Data Scientist Average Salaries by Years of Experience
df_data_scientists_filtered = df[df['Role']=='Data Scientist']
data_scientists_average_salaries_by_years = df_data_scientists_filtered.groupby(['Role', 'Years of Experience'])['Gross Salary(ksh)'].mean().astype(int).reset_index(name='Average Salary').sort_values(by=['Role','Average Salary'], ascending=[False,True])
data_scientists_average_salaries_by_years 
```

```python
#Data Engineers Average Salaries by Years of Experience
df_data_engineers_filtered = df[df['Role']=='Data Engineer']
data_engineers_average_salaries_by_years = df_data_engineers_filtered.groupby(['Role', 'Years of Experience'])['Gross Salary(ksh)'].mean().astype(int).reset_index(name='Average Salary').sort_values(by=['Role','Average Salary'], ascending=[False,True])
data_engineers_average_salaries_by_years 
```

```python
#Work setup for data professional and respective percentage.
Work_setup = df_filtered.groupby('Work Setup')['Work Setup'].count().reset_index(name='Number')

# Calculate the percentage
Work_setup['Percentage'] = round((Work_setup['Number'] / Work_setup['Number'].sum()) * 100,2)
Work_setup.drop(columns= 'Number', inplace=True)
Work_setup
```

```python
#Number of data professionals by employer type
employer_type=df_filtered.groupby('Employer Type')['Employer Type'].count().reset_index(name='Number')

#Calculating Percentage
employer_type['Percentage']=  round(employer_type['Number']/employer_type['Number'].sum()*100,2)
employer_type.drop(columns='Number', inplace=True)
employer_type
```

```python
# Most popular Tech Stack
# Spliting each 'Tech Stacks' entry into individual stacks and flattening the list
all_tech_stacks = df['Tech Stacks'].str.split(', ').sum()

# Removing any leading or trailing whitespace from each tech stack entry
all_tech_stacks_clean = [stack.strip() for stack in all_tech_stacks]

# Counting the occurrences of each tech stack
tech_stack_counts = Counter(all_tech_stacks_clean)

# Convert to DataFrame
tech_stack_df = pd.DataFrame(tech_stack_counts.items(), columns=['Tech Stacks', 'Count'])
tech_stack_df = tech_stack_df.sort_values(by='Count', ascending=False).reset_index(drop=True)
#Dropping a row with null values
tech_stack_df=tech_stack_df.drop(10)
#Percentage
tech_stack_df['Percentage']= round(tech_stack_df['Count']/ tech_stack_df['Count'].sum()*100,2)
tech_stack_df.drop(columns= 'Count', inplace=True)
tech_stack_df
```
